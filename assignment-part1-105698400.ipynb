# Data Loading and Initial Inspection

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the datasets
dropout_df = pd.read_csv('data_dropout_a.csv')
enrolled_df = pd.read_csv('data_enrolled_a.csv')
graduate_df = pd.read_csv('data_graduate_a.csv')
inflation_df = pd.read_csv('inflation_copy.csv')

# Display dataset information
print("=== Dropout Dataset Info ===")
display(dropout_df.info())
display(dropout_df.head())

print("\n=== Enrolled Dataset Info ===")
display(enrolled_df.info())
display(enrolled_df.head())

print("=== Graduate Dataset Info ===")
display(graduate_df.info())
display(graduate_df.head())

```

# Data Integration

```python
# Add a 'Target' column to indicate dropout vs enrolled status
dropout_df['Target'] = 'Dropout'
enrolled_df['Target'] = 'Enrolled'
graduate_df['Target'] = 'Graduate'

# Combine the datasets
combined_df = pd.concat([dropout_df, enrolled_df, graduate_df], ignore_index=True)

# Verify the combination
print("\nCombined Dataset Info:")
print(combined_df.info())
print("\nValue counts for Target:")
print(combined_df['Target'].value_counts())
```

# Data Cleaning

## Check for Missing Values

```python
# Check for missing values
print("\nMissing values per column:")
print(combined_df.isnull().sum())

# Calculate percentage of missing values
missing_percent = combined_df.isnull().sum() / len(combined_df) * 100
print("\nPercentage of missing values:")
print(missing_percent[missing_percent > 0])
```

## Handle Missing Values

```python
# Check for missing values
missing_values = combined_df.isnull().sum()
missing_percent = (missing_values / len(combined_df)) * 100
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percent})
display(missing_data[missing_values > 0].sort_values('Percentage', ascending=False))

# Drop columns with >50% missing values if any
combined_df = combined_df.dropna(thresh=len(combined_df)*0.5, axis=1)

# Impute remaining missing values
# Numerical columns - median imputation
num_cols = combined_df.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
    if combined_df[col].isnull().sum() > 0:
        combined_df[col].fillna(combined_df[col].median(), inplace=True)

# Categorical columns - mode imputation
cat_cols = combined_df.select_dtypes(include=['object']).columns
for col in cat_cols:
    if combined_df[col].isnull().sum() > 0:
        combined_df[col].fillna(combined_df[col].mode()[0], inplace=True)

# Verify no missing values remain
print("\nRemaining missing values after cleaning:", combined_df.isnull().sum().sum())
```

## Check for Duplicates

```python
# Check for duplicates
print("\nNumber of duplicates:", combined_df.duplicated().sum())

# Remove duplicates if any
combined_df = combined_df.drop_duplicates()
print("Number of duplicates after cleaning:", combined_df.duplicated().sum())
```

## Clean Inconsistent Values

```python
# Clean 'Gender' column
combined_df['Gender'] = combined_df['Gender'].str.lower().str.strip()
gender_mapping = {
    'female ': 'female',
    'femaLe': 'female',
    'fem': 'female',
    'male.': 'male',
    'mal': 'male',
    'maLe': 'male',
    'MALE': 'male',
    'FEMALE': 'female'
}
combined_df['Gender'] = combined_df['Gender'].replace(gender_mapping)

# Clean 'Martial status' column
combined_df['Martial status'] = combined_df['Martial status'].str.lower().str.strip()
martial_mapping = {
    'singel': 'single',
    'legally seperated': 'legally separated'
}
combined_df['Martial status'] = combined_df['Martial status'].replace(martial_mapping)

# Verify cleaning
print("\nGender distribution after cleaning:")
display(combined_df['Gender'].value_counts())

print("\nMartial status distribution after cleaning:")
display(combined_df['Martial status'].value_counts())
```

### 2.2.2 Relationship: Tuition Fees Up-to-Date vs Dropout Rate (with Aggregation)


```python
# Create aggregated data
fee_status = df.groupby(['Debtor', 'Target']).size().unstack()
fee_status['Dropout Rate'] = fee_status['Dropout'] / (fee_status['Dropout'] + fee_status['Enrolled']+fee_status['Graduate']) * 100

plt.figure(figsize=(8, 5))
sns.barplot(data=fee_status.reset_index(), x='Debtor', y='Dropout Rate')
plt.title('Dropout Rate by Tuition Fee Payment Status', fontsize=14)
plt.xlabel('Debtor (1=Yes, 0=No)', fontsize=12)
plt.ylabel('Dropout Rate (%)', fontsize=12)

# Add value labels
for index, value in enumerate(fee_status['Dropout Rate']):
    plt.text(index, value + 1, f"{value:.1f}%", ha='center')

plt.show()

# Statistical test (Chi-square)
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Debtor'], df['Target'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"\nChi-square test results: p-value = {p:.4f}")
print("The relationship is statistically significant." if p < 0.05 else "No significant relationship found.")
print(df['Debtor'].value_counts(normalize=True) * 100)
```

## 2.3 Generate and Test a Hypothesis

### Hypothesis: Older students have a higher dropout rate than younger students

```python
# Create age groups
df['Age Group'] = pd.cut(df['Age at enrollment'], 
                         bins=[0, 20, 25, 30, 100],
                         labels=['<20', '20-25', '25-30', '30+'])

# Calculate dropout rates by age group
age_dropout = df.groupby('Age Group')['Target'].value_counts(normalize=True).unstack() * 100
age_dropout['Dropout Rate'] = age_dropout['Dropout']

plt.figure(figsize=(10, 5))
sns.barplot(data=age_dropout.reset_index(), x='Age Group', y='Dropout Rate', order=['<20', '20-25', '25-30', '30+'])
plt.title('Dropout Rate by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Dropout Rate (%)', fontsize=12)

# Add value labels
for index, value in enumerate(age_dropout['Dropout Rate']):
    plt.text(index, value + 1, f"{value:.1f}%", ha='center')

plt.show()

# Statistical test (ANOVA)
from scipy.stats import f_oneway
groups = [group['Age at enrollment'].values for name, group in df.groupby('Age Group')]
f_stat, p_value = f_oneway(*groups)
print(f"\nANOVA results: p-value = {p_value:.4f}")
print("There is a significant difference between age groups." 
      if p_value < 0.05 else "No significant difference between age groups.")
```
